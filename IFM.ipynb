```python
import pandas as pd
import os
from geopy.geocoders import Nominatim
import geocoder
from geopy.extra.rate_limiter import RateLimiter
```

```python
#To merge all files into one dataframe automatically
def lister_fichiers(repertoire):
    liste_fichiers = []
    for filename in os.listdir(repertoire):
        if filename.endswith('txt'):
            liste_fichiers.append(filename)
    return liste_fichiers

liste_fichiers = lister_fichiers("data/IFM")

def lister_dataframes(liste_de_fichiers):
    NOM_DES_COLONNES = ["Année","Mois","Jour","Latitude","Longitude","IFM-max"]
    liste_dataframes = []
    for filename in liste_de_fichiers:
        filepath = f"data/IFM/{filename}"
        dataframe = pd.read_csv(filepath, 
                                sep=";", 
                                comment='#', 
                                names=NOM_DES_COLONNES)
        liste_dataframes.append(dataframe)
    return liste_dataframes

liste_dataframes = lister_dataframes(liste_fichiers)

df = pd.concat(liste_dataframes)

# To obtain one column with the coordinates
df["Coordonnées"] = df["Latitude"].map(str) + ", " + df["Longitude"].map(str)
df = df.drop(columns=["Latitude","Longitude"])
```
